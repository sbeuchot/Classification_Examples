# Classifying Handwriten Digits
## Using two different approaches: K-Nearest Neighbors (KNN) and Support Vector Machines (SVM)

The purpose of this notebook is to examplify how to use KNN and SVM to classify digits from images of handwriten digits.

The dataset used is `load_digits()` from `sklearn.datasets`, which is a copy of UCI ML hand-written digits datasets https://archive.ics.uci.edu/ml/datasets/Optical+Recognition+of+Handwritten+Digits.

Begin by loading the dataset from sklearn and store and store in an object called `digits`.

```python
import sklearn.datasets

# Load the digits dataset as object `digits`
digits = sklearn.datasets.load_digits()
```

### Description of the dataset: 
The data set contains 1797 images of hand-written digits: 10 classes where each class refers to a digit. Preprocessing programs made available by NIST were used to extract normalized bitmaps of handwritten digits from a preprinted form. From a total of 43 people, 30 contributed to the training set and different 13 to the test set. 32x32 bitmaps are divided into nonoverlapping blocks of 4x4 and the number of on pixels are counted in each block. This generates an input matrix of 8x8 where each element is an integer in the range 0 to 16.

displaying some of the images in the dataset:

```python
import matplotlib.pyplot as plt

# reshape on the data needs to be done

# displaying first digit from dataset
print('display first digit:', digits.target[0])
plt.imshow(digits.data[0].reshape(8,8),
           cmap='gray', 
           interpolation='nearest')
plt.show()

# displaying digit 968 from dataset
print('display digit 968:', digits.target[967])
plt.imshow(digits.data[967].reshape(8,8),
           cmap='gray', 
           interpolation='nearest')
plt.show()
```

### Splitting into train and test sets
Next step is to split the dataset into train and test set. Also splitting the labels (`digits.target`). Data is split into objects `X_train` and `X_test`, and labels into `y_train` and `y_test`. The split is made to have data to build/fit the classification model and a different dataset to validate the model by comparing the classifications from the model with the actual classes for those images.

The train set will have 70% of the data and the test set 30%. Setting the split as stratified on `y` so the train and test sets have approximately the same percentage of samples of each target class as the complete set.

```python
from sklearn.model_selection import train_test_split as split

X = digits.data
y = digits.target

# Split into training and test set
X_train, X_test, y_train, y_test = split(X, y,
                                         test_size=0.3,
                                         random_state=42,
                                         stratify=y)
```

## Building a K-Nearest Neighbors (KNN) model. 
The input for this model consists of the $ k $ closest training examples. So finding the optimal $ k $ is the most important part of this model.

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

# Setup arrays to store train and test accuracies
neighbors = np.arange(1, 11)
train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))

# Loop over different values of k
for i, k in enumerate(neighbors):
    # Build KNN Classifier with k neighbors: knn
    knn = KNeighborsClassifier(n_neighbors=k)

    # Fit the classifier to the training data
    knn.fit(X_train,y_train)
    
    #Compute accuracy on the training set
    train_accuracy[i] = knn.score(X_train, y_train)

    #Compute accuracy on the testing set
    test_accuracy[i] = knn.score(X_test, y_test)

# Generate plot
plt.title('KNN: Finding Optimal Number of Neighbors')
plt.plot(neighbors, test_accuracy, label = 'Testing Accuracy')
plt.plot(neighbors, train_accuracy, label = 'Training Accuracy')
plt.legend()
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Accuracy')
plt.show()
```

From the graph $ k=5 $ seems has the best accuracy for the test set. Even if the accuracy of the training set seems to increase when reducing the number of k-nearest neighbors, there doesn't seem to be an effect on the accuracy of the test set which can be an indication of overfitting for values lower than $ k=5 $.

```python
# Create a k-NN classifier with 5 neighbors: knn
knn = KNeighborsClassifier(n_neighbors=5)

# Fit the classifier to the training data
knn.fit(X_train, y_train)

# calculating the accuracy
knn_accuracy = knn.score(X_test, y_test)

# Print the accuracy
print('The accuracy of this model is: {:0.2%}'.format(knn_accuracy),
      'meaning that {:0.2%}'.format(knn_accuracy),
      'of the data in the test set is classified correctly by the model')
```

### Using the KNN model to classify new digits

Next cell will create a digit '8' to use as an example. And the KNN model will be used to classify it.

```python
#setting a random seed so it can be replicable
import random
random.seed(20)

# creating a digit 8
new_digit = np.array([[0,0,random.randint(1,17),random.randint(1,17),random.randint(1,17),random.randint(1,17),0,0,
                       0,0,random.randint(1,17),0,0,random.randint(1,17),0,0,
                       0,0,random.randint(1,17),0,0,random.randint(1,17),0,0,
                       0,0,random.randint(1,17),random.randint(1,17),random.randint(1,17),random.randint(1,17),0,0,
                       0,0,random.randint(1,17),random.randint(1,17),random.randint(1,17),random.randint(1,17),0,0,
                       0,0,random.randint(1,17),0,0,random.randint(1,17),0,0,
                       0,0,random.randint(1,17),0,0,random.randint(1,17),0,0,
                       0,0,random.randint(1,17),random.randint(1,17),random.randint(1,17),random.randint(1,17),0,0]])

# displaying the new digit
print('displaying the new created digit:')
plt.imshow(new_digit.reshape(8,8),
           cmap='gray',
           interpolation='nearest')
plt.show()

# using KNN model to classify new digit
knn_new_digit_class = knn.predict(new_digit)
print('model classify new digit as: {}'.format(knn_new_digit_class))
```

The KNN model classified the new digit as an '8', this is a correct classification.

## Building a Support Vector Machine (SVM)

Two hyperparameters that needs tunning in SVM are `C` and `gamma`. Parameter `C` determines the amount of regularization, the lower the value the higher the regularization. Regularization is use to decrease risk of overfitting and reduce time of convergence of the model. Parameter `gamma` can be interpreted as the influence a record has over the rest, low values means more influence and high values meaning less.

In the next cell instead of a grid search to find the optimal parameters, a randomized search is used to reduce the computational time and get good approximations of the optimal parameters.

```python
from sklearn.svm import SVC
from sklearn.model_selection import RandomizedSearchCV

# building a SVM model
svm = SVC(random_state=42)

# parameters to optimize are 'C' and 'gamma'
parameters = {'C':np.arange(0.01, 1, 0.01),
              'gamma':np.arange(0.00001, 0.1, 0.00001)}

# using cross-validation to find optimal parameters
param_search = RandomizedSearchCV(svm, parameters,
                                  cv=5, iid=False,
                                  random_state=42)
param_search.fit(X_train,y_train)

# displaying the best parameters
print("Best parameters", param_search.best_params_)
```

Next cell will fit the train data to the svm with the parameters found in the cell above.

```python
# builing the model with the parameters find from the CV search
svm = param_search.best_estimator_

# fitting the train set to the model
svm.fit(X_train, y_train)

#calculating the accuracy of SVM model
svm_accuracy = svm.score(X_test, y_test)

# display the accuracy
print('The accuracy of this model is: {:0.2%}'.format(svm_accuracy),
      'meaning that {:0.2%}'.format(svm_accuracy),
      'of the data in the test set is classified correctly by the model')
```

### Using the SVM model to classify new digits

We'll use the same example as with the KNN model, The digit 8 created before.

```python
# using KNN model to classify new digit
svm_new_digit_class = svm.predict(new_digit)
print('model classify new digit as: {}'.format(svm_new_digit_class))
```

The SVM model classifies the digit as 8 which is a right classification.

## Comparison of the models

A way to compare models is by the accuracy each of them has. The KNN model has an accuracy of 98.7%, a bit better than the SVM model which has 94.07%. Both accuracies are considered good, so either model can be used to classify new handwritten digits which had been formated in the same way as the dataset used to build both models.
