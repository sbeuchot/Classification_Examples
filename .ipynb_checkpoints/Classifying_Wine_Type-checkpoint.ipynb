# Wine Type Classification

## Using two different approaches: Multinomial Logistic Regression and Decision Trees.

The purpose of this notebook is to examplify how to model a multinomial linear regressions and decision trees, and use them to predict the class of a wine type depending on a set of independent variables.

The dataset used is `load_wine()` and is taken from `sklearn.datasets`, which is a copy of UCI ML Wine recognition datasets. https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data

Begin by loading the dataset and see some of the statistics of the dataset like mean, standard deviation, minimum and maximum values for each of the predictors. Also check for missing values, which in this case there are none.

```python
import sklearn.datasets

wine = sklearn.datasets.load_wine()
wine.DESCR
```

The predictors are 13 numeric attributes: 'Alcohol', 'Ash', 'Magnesium', 'Flavonoids', 'Proanthocyanins', 'Hue', 'Proline', 'Malic Acid', 'Alcalinity of Ash', 'Total Phenols', 'Nonflavonoid Phenols', and 'OD280/OD315 of diluted wines'. The target variable is categorical with three classes representing three different cultivators: `'Class_0'`, `'Class_1'`, and `'Class_2'`.

This dataset has 178 records, distributed as follow: 59 as `'Class_0'`, 71 as `'Class_1'`, and 48 as `'Class_2'`. 

Storing the predictors into an object named `X` and the target values in object `y` to facilitate the analysis.

```python
X,y = wine.data, wine.target
```

### Exploratory Data Analysis (EDA)
Doing some EDA to get familiar with the data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.DataFrame(X, columns=wine.feature_names)
pd.plotting.scatter_matrix(df, c=y, figsize=[12,12],
                           s=10, marker='D', cmap='Accent')

# Creating a lengend for the plot
handles = [plt.plot([],[], color=plt.cm.Accent(i/2.),ls='', marker='D', \
                    markersize=np.sqrt(10))[0] for i in range(3)]
labels=["Class_0", "Class_1", "Class_2"]
plt.legend(handles, labels, loc=(1.02,0))
plt.show()
```

Some easy insights from the chart: `Class_0` has high proline, `Class_1` has low color intensity, and `Class_2` has low hue and od280/od315 of diluted wines.

###  Splitting into train and test datasets.
Next step is to split both objects into train and test sets. Object `X` is split into objects `X_train` and `X_test`, and object `y` is split into `y_train` and `y_test`. The split is made to have data to build/fit the classification model and a different dataset to validate the model by comparing the predictions with the actual classes for those records.

For this example the train set will have 70% of tedata and the test set 30%. Also by stratifying in `y` bothsets will have apprximately the same percentage of sample o each target class as the complete dataset.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3,
                                                    stratify=y,
                                                    random_state=84)
```

## Building a Logistic Regression. 
In multinomial (also called softmax) approach, the loss minimised is the multinomial loss fit across the entire probability distribution instead of fitting a binary problem for each label.

Since this dataset's response variable is multiclass, some hyperparameters of the funtion `LogisticRegression` need to be changed. `multi_class` needs to be changed to `'multinomial'` and `solver` can be changed to either of `'newton-cg'`, `'sag'`, `'saga'` or `'lbfgs'`. 

Standarization is not entirely necessary in logistic regression, but in this case Ridge penalty regularization (hyperparameter `l2`) is being used, since Ridge regularization penalices on the magnitud of the coefficients associated to each variable, standardizing will provide a more fairly penalization. Another hyperparameter that needs tunning is `C`. This parameter determines the amount of regularization, the lower the value the higher the regularization. Regularization is use to decrease risk of overfitting and reduce time of convergence of the model.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler

# building the scaler to standardize the data
scaler = StandardScaler()

# standardizing the data
standardized_X_train = scaler.fit_transform(X_train)

# printing the new mean for the attributes
for i in range(13):
    print('Mean of standardized attribute',
          i, ':',
          round(standardized_X_train[:,i].mean(),5), ' ',
         'standard deviation of standardized attribute',
          i, ':',
          round(standardized_X_train[:,i].std(),5))
```

After Standardizing the mean of the attributes is zero and the standard deviation is one.

A grid search can be used to find the optimal value for hyperparameter `'C'`. As explained above, this hyperparameter determines the regularization of the data in the logistic regression model to be build. in a grid search, an exhaustive search over a list of specified values is made to find the value that optimizes the model.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

# there would be convergence warnings for every `C` value tested
# this next two lines are to stop the warnings from being printed
import warnings
warnings.filterwarnings('ignore')

# Building Linear Regression model
lr = LogisticRegression(multi_class='multinomial', solver='lbfgs',
                       penalty='l2')

# finding best value of hyperparameter C
best_C = GridSearchCV(lr, {'C':np.arange(0.01, 1, 0.01)}, cv=5)
best_C.fit(standardized_X_train, y_train)
print('the best value for hyperparameter C is: ', best_C.best_params_['C'])
```

A pipeline can be made to execute several steps in a single code. In this case a pipeline will be used to execute the standarization with the object `scaler` created a couple code cells above and the logistic regression object. The final logistic regression object is going to be built including the best parameter `'C'` calculated above.

```python
from sklearn.pipeline import Pipeline

# Rebuilding Linear Regression model with best hyperparameter C
lr = LogisticRegression(multi_class='multinomial', solver='lbfgs',
                       C=best_C.best_params_['C'])

# building pipeline
pipeline = Pipeline([('standarization', scaler),('logistic_regression', lr)])
```

Now that the pipeline object has been built, it needs to be fitted with the train dataset to have a functioning logistic regression model.

```python
# fit the train dataset
pipeline.fit(X_train, y_train)

# printing the coefficients of the model
print('coefficients to calculate probabilities for Class_0:',
      lr.coef_[0])
print('coefficients to calculate probabilities for Class_1:',
      lr.coef_[1])
print('coefficients to calculate probabilities for Class_2:',
      lr.coef_[2])
```

In a multinomial linear regression there are the same number of set of coefficients as there are class types, in this case the model has three set of coefficients, one for each class type. Coefficients are used in the following formulas to calculate the probability of a record being classify as the class those coefficents correspond to.

$ P(\hat{y}_i=c) = \dfrac{e^{\left( \hat{\beta}_{c1} x_1+\hat{\beta}_{c2} x_2+ \dots +\hat{\beta}_{cn} x_n\right) }}{\Sigma^K_{k=1}e^{\hat{\beta}_{k1} x_1+\hat{\beta}_{k2} x_2+ \dots +\hat{\beta}_{kn} x_n}} $

where c is the specifc class that the proablity is being calculated and K is the total amount of different class types.

$ P(\hat{y}_i=c) = softmax(c,\hat{\beta}_{1,1} x_1+\hat{\beta}_{1,2} x_2+ \dots +\hat{\beta}_{1,n} x_n,
\hat{\beta}_{2,1} x_1+\hat{\beta}_{2,2} x_2+ \dots +\hat{\beta}_{2,n} x_n,
\dots,
\hat{\beta}_{K,1} x_1+\hat{\beta}_{K,2} x_2+ \dots +\hat{\beta}_{K,n} x_n) $

Where the softmax function serves as the equivalent of the logistic function in binary logistic regression.

Now that a model has been built and fitted with a train dataset, predictions can be made. A prediction can be made for any record with the attributes used to fit the model, including records in the train set. But predicting records used to build the model is redundant and doesn't prove the efficiency of the model.

This is where the test dataset is useful, predicting on the test dataset and comparing the with the actual classification is the way to calculate the accuracy of a model. This metric is the fraction of right predictions the model had. A Thing to consider when using accuracy as an evaluation metric is the proportion of the different classes in the original dataset. If one of the classes has a proportion of 90% in the dataset, the model can classify all the records as that class and have a 90% accuracy, however this is not useful when identifying the other classes is also important. The proportion of different classes in this dataset is fairly even distributed so in this case accuracy is a relevant metric to use.

```python
# computing the accuracy
lr_accuracy = pipeline.score(X_test, y_test)

print('The accuracy of this model is: {:0.2%}'.format(lr_accuracy),
      'meaning that {:0.2%}'.format(lr_accuracy),
      'of the data in the test set is predicted correctly by the model')
```

### Predicting probabilities. 
In this case the prediction will be done on the test dataset to exemplify, but of course predictions are meant to be done to classify new data where the response is unknown.

Also predicting probabilities allows to make more profund analysis on the certainty of the predictions of the model.

```python
# predicting probabilities for the test dataset
lr_y_pred_proba = pipeline.predict_proba(X_test)

# showing predicted probabilities for first record in test set
print('predicted probabilities for first record in test set:',
      lr_y_pred_proba[0])

# showing actual class of first record in test set
print('actual class for first record in test set:',
      y_test[0])
```

First record has a probability to be classified as `Class_0` of 0.2%, as `Class_1` of 99.5%, and as `Class_2` of 0.3%.
Based on the probabilities the first record in the test set would be classified as `Class_1`. And the actual class for that record is `Class_1`, so in this case the model is right in its prediction. Also the model was really certain about its prediction since it had 99.5% probability of being that class.

The model can also predict the class instead of the probabilities, but with a default cutoff point of having the greatest probability. Sometimes is better to set the cutoff point based on the specifications of the problem that is being tackled.

```python
# predicting classes for the test dataset
lr_y_pred = pipeline.predict(X_test)

# predicting the class of the first record in test set
print('predicted class for first record in test set:',
      lr_y_pred[0])

# showing actual class of first record in test set
print('actual class for first record in test set:',
      y_test[0])
```

the predicted class for the first record of the test set is `Class_1`. As shown previously, the probability of it being `Class_1` was higher than the other two classes so the model classifyies this record as `Class_1`.

## Building a Decision Tree

When building a decision tree, the purpose is to split the data into branches (or paths) until the records of the same class are bundled together. So that when new data is passed thorugh the model it follows these branches (or path) until a final node that will give the classification to this new records.

Something to consider when building classification trees is that if the model splits into too much branches there may be an overfitting issue. This happens because when the model is being build, the data it uses to fit the model can have some predictor-target relation that occured by chance. And when using the model on new data those occurances will interfere in the classification. Basically when there is too much branching the model will perform good on the data that it used to be built, but not so good on new data. In order to find the optimal number of branches for the model cross-validation can be used to measure the accuracy of the model at different tree sizes (tree size = number of final nodes/unique paths), then compare these accuracies to find the best tree size.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

# building a decision tree model
tree_model = DecisionTreeClassifier(random_state=24)
tree_model.fit(X_train, y_train)

# Using cross-validation to find optimal tree size
best_tree_size = GridSearchCV(tree_model,
                              {'max_depth':range(3,50)},
                              cv=3)

best_tree_size.fit(X_train, y_train)
print ('Optimal tree size:',
       best_tree_size.best_params_['max_depth'])
```

For this case the optimal tree size is four, now this parameter has to be fit into the model.

Next cell fits the train set to the decision tree model with the optimal tree size, and then calculates the accuracy.

```python
# assign optimal tree size to tree_model
tree_model = best_tree_size.best_estimator_

# fitting model with train set
tree_model.fit(X_train, y_train)

# calculating the accuracy of this model
tree_accuracy = tree_model.score(X_test, y_test)

print('The accuracy of this model is: {:0.2%}'.format(tree_accuracy),
      'meaning that {:0.2%}'.format(tree_accuracy),
      'of the data in the test set is predicted correctly by the model')
```

### Predicting probabilities. 
Same as when using logistic regression, decision tree can predict the probabilities of the records being classified as each target label.

probabilities depend on proportion each class has on the final node the new record lands on.

```python
# predicting probabilities for the test dataset
tree_y_pred_proba = tree_model.predict_proba(X_test)

# showing predicted probabilities for first record in test set
print('predicted probabilities for first record in test set:',
      tree_y_pred_proba[0])

# showing actual class of first record in test set
print('actual class of first record in test set:',
      y_test[0])

# showing predicted probabilities for record 20 in test set
print('predicted probabilities for record 20 in test set:',
      tree_y_pred_proba[19])

# showing actual class of record 20 in test set
print('actual class of record 20 in test set:',
      y_test[19])
```

The examples above show that the first record in the test set has 100% probability of being classified as `class_1` by the decision tree model. This is because this node of the tree had only records classified as `class_1` when being fitted by the train set. In contrast, record 20 of the test set is given $ \frac{1}{3} $ probability of being `Class_0`and $ \frac{2}{3} $ of being `Class_1`. Again this is because the node where record 20 ended had those proportions when being fed the train set.

Decision tree model can also return the predicted class of a record instead of the probabilities.

```python
# predicting classes for the test dataset
tree_y_pred = tree_model.predict(X_test)

# predicting the class of the first record in test set
print('predicted class for first record in test set:',
      tree_y_pred[0])

# showing actual class of first record in test set
print('actual class for first record in test set:',
      y_test[0])
```

Again the prediction for the first record in the test set is correct.

## Comparison of the models

A way to compare models is by the accuracy each of them has. The logistic regression model has an accuracy of 98.15%, a bit better than the decision tree model which has 94.44%. Both accuracies are considered good, so either model can be used to classify or predict the classes of wine type into `Class_0`, `Class_1`, and `Class_2` using the same predictor as in the dataset used here.
